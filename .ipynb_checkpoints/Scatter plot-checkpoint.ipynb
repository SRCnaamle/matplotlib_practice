# Scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np

day=list(np.arange(1,6))
sales=np.random.randint(300,1000,len(day)).tolist()
```

```python
sizes=[300,200,400,350,260]
clr=np.array(sizes) / 100
clr=clr.tolist()
plt.scatter(day,sales,c=clr,s=sizes,marker='*',cmap='plasma')
t=plt.colorbar()
t.set_label("colorbar",fontsize=15)
plt.xlabel("days",fontsize=15)
plt.ylabel("sales",fontsize=15)
plt.title("sales on every days",fontsize=20)

plt.show()
```

```python
sales2=[300,200,400,350,260]
sizes=[300,200,400,350,260]
clr=np.array(sizes) / 100
clr=clr.tolist()
plt.scatter(day,sales,c=clr,s=sizes,marker='*',cmap='plasma',label="sales1")
plt.scatter(day,sales2,c='g',label="sales2")
t=plt.colorbar()
t.set_label("colorbar",fontsize=15)
plt.xlabel("days",fontsize=15)
plt.ylabel("sales",fontsize=15)
plt.title("sales on every days",fontsize=20)
plt.legend()

plt.show()
```

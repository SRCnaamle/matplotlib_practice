##bar plot##

```python
import matplotlib.pyplot as plt
x=["python",'c','c++','java'] # you can use strings or numbers as the x axis but y axis must be number as is specifies the height of perticular value in the x axis
y=[45,50,34,24]
plt.bar(x,y)
plt.show()
```

```python
x=["python",'c','c++','java']
y=[45,50,34,24]
plt.xlabel("languages")
plt.ylabel("popularity out of 100")
plt.title("popularity of different programming languages")
plt.bar(x,y)
plt.show() 
```

```python
x=["python",'c','c++','java']
y=[45,50,34,24]
plt.xlabel("languages",fontsize=20)
plt.ylabel("popularity out of 100",fontsize=20)
plt.title("popularity of different programming languages",fontsize=30)
plt.bar(x,y)
plt.show() 
```

```python
x=["python",'c','c++','java']
y=[45,50,34,24]
plt.xlabel("languages",fontsize=20)
plt.ylabel("popularity out of 100",fontsize=20)
plt.title("popularity of different programming languages",fontsize=30)
c=['r','m'] #you can give color according to you. 
plt.bar(x,y,width=0.2,color=c) #also you can give width to each bar using width after x and height variable
plt.show()  
```

```python
x=["python",'c','c++','java']
y=[45,50,34,24]
plt.xlabel("languages",fontsize=20)
plt.ylabel("popularity out of 100",fontsize=20)
plt.title("popularity of different programming languages",fontsize=30)
c=['g','m']
plt.bar(x,y,width=0.2,color=c,edgecolor='y',linewidth=4,linestyle=':') #edgecolor for edge and linewidth for width of edge and edgestyle to give style to line
plt.show()  
```

```python
x=["python",'c','c++','java']
y=[45,50,34,24]
plt.xlabel("languages",fontsize=10)
plt.ylabel("popularity out of 100",fontsize=10)
plt.title("popularity of different programming languages",fontsize=20)
c=['g']
plt.bar(x,y,width=0.2,color=c,edgecolor='y',linewidth=4,linestyle=':',alpha=0.5,label="popularity")
plt.legend()
plt.show()    
```

```python
x=["python",'c','c++','java']
y=[45,50,34,24]
z=[50,80,76,47]
k=[88,34,89,44]

plt.xlabel("languages",fontsize=10)
plt.ylabel("popularity out of 100",fontsize=10)
plt.title("popularity of different programming languages",fontsize=20)

import numpy as np
p=np.arange(len(x))
width=0.2

plt.bar(p,y,width,color='y',label="popularity")   # p is the position for bars in this plot
plt.bar(p+width,z,width,color='m',label="engagement")   # p + width is the position for bars in this plo
plt.bar(p+width*2,z,width,color='g',label="income")  # p + 2*width is the position for bars in this plot

plt.xticks(p+width,x,rotation=20)          #xticks puts x values at (p+width/2) positions in x axis and rotates it by 20 degree
plt.legend()
plt.show()    
```

```python
x=["python",'c','c++','java']
y=[45,50,34,24]
z=[50,80,76,47]
k=[88,34,89,44]

plt.ylabel("languages",fontsize=10)
plt.xlabel("popularity out of 100",fontsize=10)
plt.title("popularity of different programming languages",fontsize=20)

import numpy as np
p=np.arange(len(x))
width=0.2

plt.barh(p,y,width,color='y',label="popularity")   # p is the position for bars in this plot
plt.barh(p+width,z,width,color='m',label="engagement")   # p + width is the position for bars in this plo
plt.barh(p+width*2,z,width,color='g',label="income")  # p + 2*width is the position for bars in this plot

plt.yticks(p+width,x,rotation=20)          #xticks puts x values at (p+width/2) positions in x axis and rotates it by 20 degree
plt.legend()
plt.show()    
```
